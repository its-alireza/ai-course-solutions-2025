import matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = ['farvardin', 'ordibehesht', 'khordad', 'tir', 'mordad', 'shahrivar', 'mehr', 'aban', 'azar', 'dey', 'bahman', 'esfand']
temps = [12, 17, 23, 29, 31, 30, 25, 18, 11, 6, 7, 10]

plt.figure(figsize=(10,6))
plt.plot(months, temps, marker='o')
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.ylabel("Temperature (°C)")
plt.grid(alpha=1)

for i, t in enumerate(temps):
    plt.text(i, t + 1, str(t), ha='center', fontsize=10)

plt.show()
```

```python
years = list(range(2020, 2025))
ml = [94, 110, 90, 140, 125]
nlp = [80, 96, 100, 127, 122]
cv = [62, 102, 65, 110, 150]

width = 0.25
x_indexes = np.arange(len(years))

plt.figure(figsize=(10, 6))
plt.bar(x_indexes - width, ml,  width=width, label="machine learning")
plt.bar(x_indexes , nlp, width=width, label="natural language processing")
plt.bar(x_indexes + width, cv, width=width, label="computer vision")

plt.title("number of students in each field")
plt.xlabel("year")
plt.ylabel("number of Students")
plt.xticks(x_indexes, years)
plt.legend()
plt.show()
```
