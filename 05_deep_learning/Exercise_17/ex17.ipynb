```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("breast-cancer.csv")

X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
acc_kmeans = max(
    accuracy_score(y, clusters),
    accuracy_score(y, 1 - clusters)
)

print("KMeans Accuracy:", acc_kmeans)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3, metric='manhattan')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

acc_knn = accuracy_score(y_test, y_pred)
print("KNN Accuracy:", acc_knn)
```

```python
x = np.array([[0.3, 0.5, 0.8]])

# model
model = Sequential([
    Dense(3, activation='relu', input_shape=(3,)),
    Dense(2, activation='tanh'),
    Dense(1, activation='sigmoid')
])

# set w and b
model.layers[0].set_weights([
    np.array([[0.1, 0.4, 0.2],
              [0.2, 0.1, 0.3],
              [0.2, 0.2, 0.1]]),
    np.array([0.2, 0.2, 0.1])
])

model.layers[1].set_weights([
    np.array([[0.1, 0.4],
              [0.2, 0.1],
              [0.2, 0.2]]),
    np.array([0.1, 0.4])
])

model.layers[2].set_weights([
    np.array([[0.5],
              [0.1]]),
    np.array([0.5])
])

# Feed Forward
o1 = model.layers[0](x)
o2 = model.layers[1](o1)
o3 = model.layers[2](o2)

print("O11 =", o1.numpy()[0,0])
print("O22 =", o2.numpy()[0,1])
print("O31 =", o3.numpy()[0,0])
```

```python
x = np.array([[1.3], [2.1], [7.6]])
y = np.array([[2.7], [4.5], [14.9]])

w = 1.5
b = 0.5
alpha = 0.1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = w * x + b
y_pred = sigmoid(z)

# MSE gradient
dw = np.mean(2 * (y_pred - y) * y_pred * (1 - y_pred) * x)
db = np.mean(2 * (y_pred - y) * y_pred * (1 - y_pred))

w_new = w - alpha * dw
b_new = b - alpha * db

print("updated weight:", w_new)
print("updated bias:", b_new)
```

```python
def relu(z):
    return np.maximum(0, z)

def tanh(z):
    return np.tanh(z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


x = np.array([
    [0.3],
    [0.5],
    [0.8]
])

W1 = np.array([
    [0.1,0.2,0.2],
    [0.4,0.1,0.2],
    [0.2,0.3,0.1]
])

b1 = np.array([
    [0.2],
    [0.2],
    [0.1]
])

W2 = np.array([
    [0.1,0.2,0.2],
    [0.4,0.1,0.2]
])

b2 = np.array([
    [0.1],
    [0.4]
])

W3 = np.array([[0.5,0.1]])

b3 = np.array([[0.5]])


# Layer 1
o1 = relu(W1 @ x + b1)

# Layer 2
o2 = tanh(W2 @ o1 + b2)

# Output Layer
o3 = sigmoid(W3 @ o2 + b3)

print("O11 =", o1[0,0])
print("O22 =", o2[1,0])
print("O31 =", o3[0,0])
```
